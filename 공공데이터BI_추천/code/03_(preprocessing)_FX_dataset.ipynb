# 기간별평균환율 합치기

```python
import os
import numpy as np
import pandas as pd
from IPython.display import display
from tqdm import tqdm
tqdm.pandas()

import matplotlib.pyplot as plt
import seaborn as sns
plt.rc('font', family='AppleGothic')
plt.rc('axes', unicode_minus=False)

import warnings
warnings.filterwarnings('ignore')
```

# FX

```python
os.listdir('../data/기간별평균환율')
```

```python
csv_files = sorted([file for file in os.listdir('../data/기간별평균환율') if file.endswith('.csv')])
csv_files
```

```python
# 첫번째 파일 확인
file = csv_files[0]
df = pd.read_csv('../data/기간별평균환율'+'/'+file,
                skiprows=3,usecols=[0,6], names=['currency','year']) 
df.rename(columns={'year':file.split('.')[0][-4:]}, inplace=True)
df
```

```python
for file in tqdm(csv_files[1:]):
    if file.endswith('.csv'):
        try:
            data = pd.read_csv('../data/기간별평균환율'+'/'+ file, 
                               skiprows=3, usecols=[0,6], names=['currency','year']) 
            data.rename(columns={'year':file.split('.')[0][-4:]}, inplace=True)
            df = pd.merge(df, data, on='currency', how='outer')
        except Exception as e:
            print(f"Error reading file: {file}. Error message: {str(e)}")
df
```

```python
df.to_csv('../data/기간별평균환율.csv', index=False)
```
