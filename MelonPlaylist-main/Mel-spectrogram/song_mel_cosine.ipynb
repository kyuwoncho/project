```python
import os
import numpy as np
import pandas as pd
import pickle
import sklearn
```

```python
# song_mel_cosine.csv 불러오기
song_mel = pd.read_csv('/home/k0c/song_mel.csv')
song_mel
```

```python
labels = song_mel[['song_id']]
df= song_mel.drop(columns=['song_id'])
```

```python
df_rs = pd.DataFrame(df, columns=df.columns)

df_rs
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(df_rs)   

sim_df = pd.DataFrame(similarity, index=labels.index, columns=labels.index)

sim_df
```

```python
def find_similar_songs(name, n=100):
  series = sim_df[name].sort_values(ascending=False)

  series = series.drop(name)


  return series.head(n).to_frame()

find_similar_songs(0)
```
